chương trình tính toán cho 2 số:

```python
cmd=input()
if "+" in cmd:
    cmd=list(map(int,cmd.split("+")))
    a=cmd[0]
    b=cmd[1]
    print(a+b)
elif "-" in cmd:
    cmd=list(map(int,cmd.split("-")))
    a=cmd[0]
    b=cmd[1]
    print(a-b)
elif "*" in cmd:
    cmd=list(map(int,cmd.split("*")))
    a=cmd[0]
    b=cmd[1]
    print(a*b)
elif "/" in cmd:
    cmd=list(map(int,cmd.split("/")))
    a=cmd[0]
    b=cmd[1]
    try:
        print(a/b)
    except:
        print("INVALID")
else:
    print(cmd)
```
